# Year, Genre, Publisher 결측치 크롤링으로 수집

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('vgames2.csv', index_col=0)
df
```

```python
# 데이터 확인
df.info()
```

```python
# 년도 데이터 이상수치 확인 및 담아두기
year = df.Year.unique()
temp = []
for i in year:
    if i <= 1900:
        print(i)
        temp.append(i)
```

```python
# 년도 이상수치 데이터 확인 및 게임 이름을 출시일 검색
for t in temp:
    x = df.loc[(df.Year.values == t), ['Name', 'Year']]
    print(x[:1])
```

> 년도가 1900이하로 되어있는 데이터를 확인해보니 80이상은 1900년도 이하는 2000년도 인 것을 확인할 수 있었다.

```python
# 데이터 수치 통일화를 위해서 년도 데이터 변경
df.loc[(df.Year.values <= 80), 'Year'] += 2000
df.loc[((df.Year.values >= 80) & (df.Year.values <= 100)), 'Year'] += 1900

# 변경확인
df.Year.value_counts()
```

```python
df.Platform.value_counts()
```

```python
df.loc[df.Platform == '2600', 'Platform'] = 'Atari 2600'
df.loc[df.Platform == 'Atari 2600']
```

```python
# 년도 결측치 확인
names_list = []
platform_list = []
for x, y in zip(df.loc[(df.Year.isna()), 'Name'], df.loc[(df.Year.isna()), 'Platform']):
    names_list.append(x)
    platform_list.append(y)
    
print(names_list[-3:])
print(platform_list[-3:])
```

```python
na_years_df = pd.DataFrame({"Name" : names_list, "Platform" : platform_list, "Year" : 0})
na_years_df.head()
```

```python
from selenium import webdriver
import time
from bs4 import BeautifulSoup

driver = webdriver.Chrome('C:/Users/rsj99/Desktop/datasalon-master/chrome/chromedriver.exe')


for name, platform in zip(names_list, platform_list):
    driver.get('https://www.google.com/search?q={}'.format(platform + ' ' + name + ' ' +  "release date"))
    time.sleep(3)
        
    try:
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        date = soup.select('div.Z0LcW.XcVN5d')
        year = date[0].text.split(" ")[0].replace("년", '').strip()
        na_years_df.loc[na_years_df.Name.values == name, 'Year'] = int(year)
    except:
        try:
            html = driver.page_source
            soup = BeautifulSoup(html, 'html.parser')
            date = soup.select('div.Z0LcW.XcVN5d')
            year =date[0].text.split(" ")[-1].strip()
            na_years_df.loc[na_years_df.Name.values == name, 'Year'] = int(year)
        except:
            na_years_df.loc[na_years_df.Name.values == name, 'Year'] = 0
        
driver.quit()
na_years_df.to_csv('na_years.csv', index=False)
```
